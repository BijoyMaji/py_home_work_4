```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.integrate import odeint
def model(u,t):
    x=u[0]
    y=u[1]
    dxdt=y
    dydt=-2*y-x
    return np.array([dxdt,dydt])
u0=[1,0] #initial condition
t=np.linspace(0,20,100)
sol=odeint(model,u0,t)
x=sol[:,0]
y=sol[:,1]
plt.plot(t,x)
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt
from scipy.special import legendre
N=1000
xvalue=np.linspace(-1,1,N)
for i in range(1,5):
    pn=legendre(i)
    plt.plot(xvalue,pn(xvalue),label="n="+str(i))
    plt.legend(loc='upper center')
plt.title('Legendre Polynomials')
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.integrate import odeint
def model(u,t):
    x=u[0]
    y=u[1]
    dxdt=y
    dydt=4.5-x/0.9 -0.5*y
    return np.array([dxdt,dydt])
u0=[1,0] #initial condition
t=np.linspace(0,10,100)
sol=odeint(model,u0,t)
x=sol[:,0]
y=sol[:,1]
plt.plot(t,y)
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt
from scipy.fftpack import fft
n=100
x=np.linspace(0.0,2,n)
freq=np.linspace(0.0,2,n//2)
y=np.exp(-x**2)
yf=fft(y)
ya=np.abs(yf)
plt.plot(freq,ya[:n//2])
plt.show()
```
