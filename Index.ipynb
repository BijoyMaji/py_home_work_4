```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.integrate import odeint
def model(u,t):
    x=u[0]
    y=u[1]
    dxdt=y
    dydt=-2*y-x
    return np.array([dxdt,dydt])
u0=[1,0] #initial condition
t=np.linspace(0,20,100)
sol=odeint(model,u0,t)
x=sol[:,0]
y=sol[:,1]
plt.plot(t,x)
plt.show()
```
